# RETO INTEGRADOR: Sistema de Dos Lagos

**Escenario:** Una industria vierte contaminante en dos lagos conectados.
- Lago 1: Recibe el vertido (800 mg/L)
- Lago 2: Abastece un pueblo (debe bajar a <100 mg/L en 30 días)

```maxima
/* Definición del sistema */
depends(C1, t);
depends(C2, t);

V1: 500;
V2: 1000;
Q: 100;
k1: 0.15;
k2: 0.10;

C1_0: 800;
C2_0: 50;
```

```maxima
/* Sistema de EDOs acopladas */
ecuacion_lago1: 'diff(C1, t) = -(Q/V1)*C1 - k1*C1;
ecuacion_lago2: 'diff(C2, t) = (Q/V2)*C1 - (Q/V2)*C2 - k2*C2;
```

```maxima
/* Funciones para el solver numérico */
f1(C1, C2, t) := -(Q/V1)*C1 - k1*C1;
f2(C1, C2, t) := (Q/V2)*C1 - (Q/V2)*C2 - k2*C2;
```

```maxima
/* Simulación numérica con Runge-Kutta */
lista_tiempos: makelist([i, i+0.5], i, 0, 30, 0.5);
resultados: rk([f1(C1, C2, t), f2(C1, C2, t)], [C1, C2], [C1_0, C2_0], lista_tiempos);
```

```maxima
/* Extraer columnas */
tiempo: map(lambda([x], x[1]), resultados);
C1_vals: map(lambda([x], x[2]), resultados);
C2_vals: map(lambda([x], x[3]), resultados);
```

```maxima
/* Graficar ambos lagos */
plot2d(
    [[discrete, tiempo, C1_vals], [discrete, tiempo, C2_vals]],
    [style, lines, lines],
    [color, red, blue],
    [line_width, 3, 3],
    [legend, "Lago 1 (Vertido)", "Lago 2 (Pueblo)"],
    [xlabel, "Tiempo (dias)"],
    [ylabel, "Concentracion (mg/L)"],
    [title, "Sistema de Dos Lagos"],
    [y, 0, 850]
);
```

```maxima
/* Límite normativo */
plot2d(100, [t, 0, 30], 
    [color, green], [line_width, 2],
    [legend, "Limite Normativo (100 mg/L)"]);
```
