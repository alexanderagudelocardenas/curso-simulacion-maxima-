# CLASE 1: DEL MODELO MENTAL AL EXPERIMENTO COMPUTACIONAL

**Sistema:** Un lago que recibe un contaminante y lo degrada naturalmente.

---

## BLOQUE 1: LA CONSTRUCCIÓN MENTAL

Aquí definimos las "reglas del juego" de nuestro modelo.

```maxima
/* Declaramos que la Concentración (C) depende del tiempo (t) */
depends(C, t);
```

```maxima
/* Parámetros del modelo mental (Caja Blanca) */
V: 1000;       /* Volumen del lago en m^3 */
Q: 50;         /* Caudal de salida en m^3/dia */
k: 0.2;        /* Tasa de degradación biológica (1/dia) */
```

```maxima
/* La Ecuación Diferencial: Ley de Conservación de la Masa */
/* Acumulación = Entrada (0) - Salida (Q/V * C) - Reacción (k * C) */
modelo_mental: 'diff(C, t) = -(Q/V)*C - k*C;
```

---
## BLOQUE 2: EL EXPERIMENTO COMPUTACIONAL (Simulación)

Inyectamos el modelo en el motor computacional para verlo evolucionar.

```maxima
/* Resolvemos la ecuación simbólicamente */
solucion_general: ode2(modelo_mental, C, t);
```

```maxima
/* Condición Inicial: En el día 0, hubo un derrame de 500 mg/L */
C0: 500;
simulacion: ic1(solucion_general, t=0, C=C0);
```

```maxima
/* Extraemos la función C(t) */
C_t(t) := rhs(simulacion);
```

---
## BLOQUE 3: VISUALIZACIÓN

```maxima
/* Graficamos la evolución durante 20 días */
plot2d(C_t(t), [t, 0, 20], 
    [xlabel, "Tiempo (dias)"], 
    [ylabel, "Concentracion (mg/L)"],
    [title, "Simulacion Caja Blanca: Decaimiento en el Lago"],
    [color, blue], [line_width, 3]);
```

---
## BLOQUE 4: PREGUNTA DE INGENIERÍA

¿En qué día la concentración bajará a 50 mg/L?

```maxima
/* Buscamos el tiempo donde C(t) = 50 */
dia_seguro: find_root(C_t(t) - 50, t, 10, 40);

print("El lago será seguro en el día: ", dia_seguro);
```
